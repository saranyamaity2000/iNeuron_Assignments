```python
def reduce(operation , arr):
    res = arr[0]
    for i in range(1,len(arr)):
        res = operation(res, arr[i])
    return res


arr = [23,17,40,20]

# calculate sum value 
summ = reduce(lambda a , b : a + b , arr) ; 

print(f"sum = {summ}")

#calculate max value 
maxx = reduce(lambda a , b : a if a > b else b , arr)

print(f"max value = {maxx}")
```

```python
def filter(valid , arr):
    new = []
    for ele in arr:
        if(valid(ele)):
            new.append(ele)
    return new

def prime_filter(x):
    if(x == 1):
        return False
    if(x == 2 or x == 3):
        return True
    i = 2
    while(i*i <= x):
        if(x%i == 0):
            return False
        i += 1
    return True


arr = [2,3,67,23,21,56,21,89,37,81]

filtered_arr = filter(prime_filter, arr)

print(f"new list of elements filtered by prime number : {filtered_arr}")
```
